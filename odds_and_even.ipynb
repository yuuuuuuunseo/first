```python
# 필요한 패키지 설치
!pip install ultralytics opencv-python-headless matplotlib

# 파일 업로드를 위한 라이브러리 import
from google.colab import files
import os
import shutil

# 이미지 파일 업로드
uploaded = files.upload()
filename = next(iter(uploaded))  # 업로드된 파일 이름을 가져옵니다.

# 업로드된 이미지 파일을 적절한 디렉토리로 이동
destination_dir = '/content/sample_data'

# 디렉토리가 없는 경우 생성
os.makedirs(destination_dir, exist_ok=True)

# 파일 이동
shutil.move(filename, os.path.join(destination_dir, filename))

print(f"파일 {filename}이 {destination_dir}로 이동되었습니다.")
```

```python
from ultralytics import YOLO
from IPython.display import Image, display

# Load a pretrained YOLOv8n model
model = YOLO('yolov8n.pt')

# Define path to the image file
image_path = '/content/sample_data/1.jpg'

# Run inference on the source
results = model.predict(image_path)  # list of Results objects

# Assuming results[0] contains the desired Result object
first_result = results[0]

# Save the result
saved_image_paths = first_result.save()  # Save and get the path to the saved image(s)

# Display the saved image
# If `saved_image_paths` is a list, iterate through it
if isinstance(saved_image_paths, list):
    for path in saved_image_paths:
        display(Image(filename=path))
else:
    # If it's a single path (string), display the image directly
    display(Image(filename=saved_image_paths))
```
